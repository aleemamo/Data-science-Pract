# Filter application for truth

```python
ages=[5,12,17,18,24,32,45,53,64,70]

def myFunction(x):
    if x<18:
        return False
    else:
        return True
adults=filter(myFunction,ages)

for x in adults:
    print(x)
```

#### http://www.kaggle.com/code/corex0/dice-ladder-problem-probability/notebook

```python
all_walks=[]

for i in range(10):
    random_walk=[0]
    for x in range(100):
        step=random_walk[-1]
        dice = np.random.randint(1,7)
        #print(x)
        if dice <=2:
            step=max(0, step-1)
        elif dice<=5:
            step=step+1
        else:
            step=step + np.random.randint(1,7)
            #print(step)
            
        #implememt clumsiness
        rand_prob= np.random.rand()
        if rand_prob<=0.001:
            step=0
            
        #using the build in append function
        random_walk.append(step)
        
    all_walks.append(random_walk)       
```

```python
all_walks
```

```python
np_aw=np.array(all_walks)
#plot np_aw and show
plt.plot(np_aw)
plt.show()
```

```python

##clear last plot
plt.clf()
#Transpose np_aw: np_aw_t
np_aw_t=np.transpose(np_aw)

# plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```
